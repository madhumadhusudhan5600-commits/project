```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

#MISSING VALUES ANALYSIS

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(df.isnull(),cbar=False)
plt.title("missing values Heatmap")
plt.show()
```

#data cleaning

```python
df.drop(columns=['Cabin'],inplace=True)
```

```python
df.head()
```

```python
df['Age'].fillna(df['Age'].median(),inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

univariate analysis

```python
sns.countplot(x='Survived',data=df)
plt.title("Survived Count")
plt.show()
```

bivariate analysis

```python
sns.countplot(x='Survived',hue='Sex',data=df)
plt.title("Survived by Gender")
plt.show()
```

```python
sns.countplot(x='Pclass',hue='Sex',data=df)
plt.title("Pclass by Gender")
plt.show()
```

```python
sns.countplot(x='Fare',hue='Sex',data=df)
plt.title("Fare by Gender")
plt.show()
```

```python
sns.countplot(x='Name',hue='Sex',data=df)
plt.title("Name by Gender")
plt.show()
```

multivariate *analysis* it will take 2 or more variables




```python
sns.boxplot(x='Survived',y='Age',color='red',hue='Sex',data=df)
plt.title("Age vs Survival vs Gender")
plt.show()
```

```python
sns.boxplot(x='Parch',y='Age',color='blue',hue='Sex',data=df)
plt.title("Parch vs Age vs Gender")
plt.show()
```

```python
#correlation analysis(finding relationship b/w 2 variables)
+1 strong positive
0 negative
+0 strong negative
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title("correlation Heatmap")
plt.show()

```
